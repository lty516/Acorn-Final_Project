```python
import pandas as pd
```

# 독거노인데이터

- 필요한 형태 : (기초생활 + 저소득) / (총 독거 노인)

```python
data = pd.read_csv("data/독거노인_2019.txt", sep = "\t", header =[0, 1], index_col=[2])
data.head()
```

```python
# data 중에서 기초 생활, 저소득, 총 인원수만 필요
data = data.iloc[:, [1, 5, 8, 2]]
data.head()
```

```python
# 소계와 합계는 필요없으므로 삭제
data.drop(["소계", "합계"], axis = 0, inplace=True)
data
```

```python
# 행정동이 index에 있으므로 reset
data = data.reset_index()

# column 이름 변경
data.columns = ["행정동", "구", "기초생활보장수급권자", "저소득", "독거노인"]
data
```

```python
# 보기좋은 순서대로 배치
data = data[["구", "행정동", "기초생활보장수급권자", "저소득", "독거노인"]]
data.head()
```

```python
# 0인 값이 - 로 들어가 있으니 0으로 치환
data = data.replace("-", '0')
data
```

```python
# 기초생활보장수급권자, 저소득, 독거노인 모두 string 이므로 int형으로 변환하기 위해 1,330 과 같은 형태일 때 , 삭제하기
data["기초생활보장수급권자"] = data["기초생활보장수급권자"].str.replace(",", "")
data["저소득"] = data["저소득"].str.replace(",", "")
data["독거노인"] = data["독거노인"].str.replace(",", "")

# 두 컬럼 모두 int로 변환
data["기초생활보장수급권자"] = data["기초생활보장수급권자"].astype("int")
data["저소득"] = data["저소득"].astype("int")
data["독거노인"] = data["독거노인"].astype("int")
data.head()
```

```python
# 취약계층 = 기초생활보장수급권자 + 저소득 
data["취약계층"] = data["기초생활보장수급권자"] + data["저소득"]
data.head()
```

```python
basicLevelSenior = data[["구", "행정동", "취약계층", "독거노인"]]
basicLevelSenior.head()
```

# 전체 노인 인구
- 동별 전체 노인 인구를 취약계층 컬럼에 취합할 것

```python
totalSenior = pd.read_csv("data/서울시 고령자현황 (동별) 통계 (2019_4).txt", sep="\t", header = [0, 1, 2], index_col=[2])
totalSenior.head()
```

```python
# 구, 동, 총 노인 수만 남김
totalSenior = totalSenior.iloc[:,[1, 5]]
totalSenior.head()
```

```python
# 합계, 소계 삭제
totalSenior = totalSenior.drop(["합계", "소계"], axis = 0)

# index reset
totalSenior = totalSenior.reset_index()

# 컬럼명 변환
totalSenior.columns = ["행정동", "구", "총 노인인구"]

# 컬럼 순서 변환
totalSenior = totalSenior[["구", "행정동", "총 노인인구"]]

# 총 노인인구의 comma 삭제
totalSenior["총 노인인구"] = totalSenior["총 노인인구"].str.replace(",", "")

# 총 노인인구를 int형으로 변환
totalSenior["총 노인인구"] = totalSenior["총 노인인구"].astype("int")

totalSenior.head()
```

# 취약계층,  총 노인인구 dataframe 합치기

```python
seniorDf = basicLevelSenior.merge(totalSenior)
seniorDf
```

```python
seniorDf["전체노인_취약계층"] = seniorDf["취약계층"] / seniorDf["총 노인인구"]
seniorDf
```

```python
seniorDf["독거노인_취약계층"] = seniorDf["취약계층"] / seniorDf["독거노인"]
seniorDf
```

```python
seniorDf.to_csv("data/분석용 데이터/firstClusteringDF(cp949).csv", encoding = "cp949", index = False)
```
