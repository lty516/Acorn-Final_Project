```python
# CCTV for summary
cctv_summary = pd.read_csv("summary/CCTV_행정동변환(cp949).csv" , encoding='cp949')
cctv_summary
```

```python
# groupby _ sum
cctv_summary = cctv_summary.groupby("행정동").agg({'카메라대수':sum})
cctv_summary = cctv_summary.sort_values(by=["카메라대수"],ascending=False)
cctv_summary = cctv_summary.reset_index()
cctv_summary
```

```python
# # 불필요 컬럼 삭제
# cctv_summary.drop(["관리기관명","설치목적구분","위도","경도","제공기관명"], axis=1, inplace=True)
# cctv_summary

# 컬럼명 변경
cctv_summary = cctv_summary.rename({"행정동":"dong","카메라대수":"cctv_count"}, axis="columns")
cctv_summary
```

```python
# join for summary
summary = pd.merge(summary, cctv_summary, how="left")
summary = summary.fillna(0)
summary
```

```python
# Bus station
bus_station = pd.read_csv("data/버스전처리_행정동(cp949).csv", encoding='cp949')

# groupby _ sum
bus_station = bus_station.groupby("행정동").count()
bus_station.drop(["정류소번호","X좌표","Y좌표","지번주소","구"], axis=1, inplace=True)
bus_station = bus_station.reset_index()
bus_station

# 컬럼명 변경
bus_station = bus_station.rename({"행정동":"dong", "정류소명":"bus_station"}, axis="columns")

#join for summary 
summary = pd.merge(summary, bus_station, how='left')
summary
```

```python
# 상권별로 summary
commercial_district = pd.read_csv("summary/shop_summary(utf-8).csv",encoding='utf8')
commercial_district
```

```python
# 상권별로 통합
commercial_district = commercial_district.drop(0)

commercial_district['골목상권(총)'] = commercial_district['골목상권'].astype(float)  \
                                        + commercial_district['골목상권.1'].astype(float) \
                                        + commercial_district['골목상권.2'].astype(float) \
                                        + commercial_district['골목상권.3'].astype(float)
commercial_district['발달상권(총)'] = commercial_district['발달상권'].astype(float)  \
                                        + commercial_district['발달상권.1'].astype(float) \
                                        + commercial_district['발달상권.2'].astype(float) \
                                        + commercial_district['발달상권.3'].astype(float)
commercial_district['전통시장(총)'] = commercial_district['전통시장'].astype(float)  \
                                        + commercial_district['전통시장.1'].astype(float) \
                                        + commercial_district['전통시장.2'].astype(float) \
                                        + commercial_district['전통시장.3'].astype(float)
commercial_district
```

```python
# 불피요 컬럼 삭제 
commercial_district.drop(["골목상권","골목상권.1","골목상권.2","골목상권.3",
                         "발달상권","발달상권.1","발달상권.2","발달상권.3",
                         "전통시장","전통시장.1","전통시장.2","전통시장.3"],
                        axis=1, inplace=True)
commercial_district
```

```python
# 컬럼명 변경
commercial_district = commercial_district.rename({"시군구명" : "gu","읍면동명":"dong",
                                                    "골목상권(총)":"Alley_district",
                                                    "발달상권(총)":"Development_district",
                                                    "전통시장(총)":"Traditional_district"}, axis="columns")
commercial_district
```

```python
# join for summary
summary = pd.merge(summary, commercial_district, how='left')
summary
```

```python
# NaN값 0 으로 대체 
summary = summary.fillna(0)
```

```python
# save
summary.to_csv("summaryForPredict(cp949).csv", encoding="cp949",index=False)
```
