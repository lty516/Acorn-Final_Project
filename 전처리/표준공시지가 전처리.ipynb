# 데이터 불러오기 및 기본 처리

1. 서울특별시만 남기기
2. 필요 없는 컬럼 삭제
3. 주소 컬럼 생성 및 사용한 컬럼 삭제
4. 용도지역1, 2에 주거가 포함된 row 만 남기기

```python
data = pd.read_csv("data/국토교통부_표준지공시지가_20200922.csv", encoding="cp949", sep='\t')

# 1. 서울만
data = data[data["시도명"] == "서울특별시"]
data
```

```python
## 컬럼 확인하기

data.columns
# 시도명, 시군구명, 소재지, 지목, 면적, 용도지역1, 용도지역2, 이용상황, 주위환경, 지세명, 공시지가
```

- 필요없는 컬럼 : 일련번호, 본번지, 부번지, 시군구, 읍면동리, 지번구분, 형상명, 도로교통, 지리적위치1, 지리적위치2, 방위

```python
# 2. 필요없는 컬럼 삭제
data = data.drop(['일련번호', '본번지', '부번지', '시군구', '읍면동리', '지번구분', '형상명', '도로교통', 
                  '지리적위치1', '지리적위치2', '방위'], axis = 1)
data.head()
```

```python
# 3-1. 주소 컬럼 생성 -> 시도명 + 시군구명 + 소재지
data["주소"] = data["시도명"] + " " + data["시군구명"] + " " + data["소재지"]
data.head()
```

```python
# 3-2. 주소 컬럼 생성 했으므로 시도명, 소재지 삭제
data.drop(['시도명', '소재지'], axis = 1, inplace = True)
data.head()
```

```python
data.shape
```

```python
# 4. 용도지역1, 2에 주거만 남기기

## 용도지역1에 주거가 들어았는 row
purpose1 = data[data['용도지역1'].str.contains(pat = '주거')]
## 용도지역2에 주거가 들어았는 row
purpose2 = data[(data['용도지역2'].notnull())&(data['용도지역2'].str.contains(pat = '주거'))]

## purpose1 , purpose2 Union
publishPrice = pd.merge(purpose1, purpose2, how="outer")
publishPrice
```

```python
### 최종 결과 저장해서 행정동으로 변환하기 -> 카카오 API 파일에서 행할 것
publishPrice.to_csv("data/부동산/표준공시지가_전처리1(cp949).csv", encoding="cp949", index=False)
```
