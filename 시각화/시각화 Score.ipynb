```python
%matplotlib inline
import folium
import json

import pandas as pd
import numpy as np

geo_path2 = 'json/HangJeongDong_ver20191231.geojson'
geo_str2 = json.load(open(geo_path2, encoding='utf-8'))
geo_str2

features_list = []

for dong in geo_str2["features"]:
    if dong["properties"]["adm_nm"].split()[0] == "서울특별시":
        dong["properties"]["adm_nm"] = dong["properties"]["adm_nm"].replace("·", ".")
        dong["properties"]["adm_nm"] = dong["properties"]["adm_nm"].split(" ")[2]
        
        print(dong["properties"]["adm_nm"])
        features_list.append(dong)
        
geo_str2["features"] = features_list
geo_str2
```

---

```python
# load data
score_0 = pd.read_csv("data/Pretreatment/scoring_score_0(cp949).csv", encoding='cp949')
score_0
```

```python
score_0 = score_0.sort_values("score", ascending=False)[:10]
score_0.drop(["pharmacy","bus_station","food","parking","Alley_district","Development_district","culture_space"], axis=1, inplace=True)
score_0
```

```python
score_0['label'] = 0
score_0
```

```python
# score_0
map_score_0 = folium.Map(location=[37.5502, 126.982], zoom_start=10)
map_score_0.choropleth(geo_data = geo_str2, 
               data=score_0,
               columns=['dong', 'label'],
               key_on='properties.adm_nm',
               fill_color='YlOrRd')
map_score_0
```
