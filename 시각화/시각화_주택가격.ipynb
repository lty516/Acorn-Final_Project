```python
%matplotlib inline
import folium
import json

import pandas as pd
import numpy as np

geo_path2 = 'json/HangJeongDong_ver20191231.geojson'
geo_str2 = json.load(open(geo_path2, encoding='utf-8'))
geo_str2

features_list = []

for dong in geo_str2["features"]:
    if dong["properties"]["adm_nm"].split()[0] == "서울특별시":
        dong["properties"]["adm_nm"] = dong["properties"]["adm_nm"].replace("·", ".")
        dong["properties"]["adm_nm"] = dong["properties"]["adm_nm"].split(" ")[2]
        
        print(dong["properties"]["adm_nm"])
        features_list.append(dong)
        
geo_str2["features"] = features_list
geo_str2
```

###  loaddata

```python
house_price = pd.read_csv("api/clustering_final_2(cp949).csv", encoding='cp949')
house_price
```

```python
# house_price_매매
map_house_trading = folium.Map(location=[37.5502, 126.982], zoom_start=10)
map_house_trading.choropleth(geo_data = geo_str2, 
               data=house_price,
               columns=['행정동', '매매'],
               key_on='properties.adm_nm',
               fill_color='PuRd')
map_house_trading
```

```python
# house_price_전월세
map_house_monthly = folium.Map(location=[37.5502, 126.982], zoom_start=10)
map_house_monthly.choropleth(geo_data = geo_str2, 
               data=house_price,
               columns=['행정동', '전월세'],
               key_on='properties.adm_nm',
               fill_color='PuRd')
map_house_monthly
```
