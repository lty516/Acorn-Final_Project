```python
# numpy
import numpy as np

# pandas
import pandas as pd

# Matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt             #

#seaborn
import seaborn as sns

# 한글 폰트 관리! 
import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:\\Windows\\Fonts\\malgun.ttf").get_name()
plt.rc('font',family=font_name)
mpl.rcParams["axes.unicode_minus"]=False
```

## 2019 서울특별시 병의원 현황

```python
hospital = pd.read_csv("data_csv/클러스터링용/병의원_행정동별개수_2019(cp949).csv", encoding="cp949")
hospital
```

```python
# 서울시 구별 병의원 개수

# 그래프 사이즈
a4_dims = (15, 13)

# 생성
fig, ax = plt.subplots(figsize=a4_dims)

# 데이터 지정
sns.barplot(x="구", y='병의원', data=hospital, ci=False)

# 그래프 타이틀
plt.title("2019년 서울특별시 구별 병의원 개수", fontsize='20')

# x,y 축 이름
ax.set_xlabel('행정동' ,fontsize='13')
ax.set_ylabel('약국(개)' ,fontsize='13')

# x축 라벨들 45도 삐딱하게
plt.xticks(rotation=45)

# 보여주기
plt.show()
```

```python
# 구 선택하여 행정동별 그래프 뽑아보기
gu_name = input('구 이름: ')
gu_tmp = hospital[hospital["구"]== gu_name]

# 그래프 사이즈
size = (15, 13)

# 생성
fig, ax = plt.subplots(figsize = size)

# 데이터 지정
sns.barplot(x= "행정동", y='병의원', data=gu_tmp, ci=False)

# 그래프 타이틀
plt.title("2019년 서울특별시 "+ gu_name +" 병의원 개수", fontsize='20')

# x,y 축 이름
ax.set_xlabel('행정동' ,fontsize='13')
ax.set_ylabel('약국(개)' ,fontsize='13')

# x축 라벨들 45도 삐딱하게
plt.xticks(rotation=45)

# 보여주기
plt.show()
```
