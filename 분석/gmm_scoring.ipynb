# Scoring 가즈앗!

```python
score_0 = pd.read_csv("data/분석용_데이터/score_0_ver_gmm(cp949).csv", encoding = "cp949")
score_0
```

```python
score_0 = score_0[["pharmacy", "bus_station", "food", "parking", "Alley_district", "Development_district", "culture_space"]]
score_0["pharmacy"] = pd.qcut(score_0["pharmacy"], q = 9 , labels=False,duplicates="drop")
score_0["bus_station"] = pd.qcut(score_0["bus_station"], q = 9, labels=False ,duplicates="drop")
score_0["food"] = pd.qcut(score_0["food"], q = 9, labels=False ,duplicates="drop")
score_0["parking"] = pd.qcut(score_0["parking"], q = 9, labels=False ,duplicates="drop")
score_0["Alley_district"] = pd.qcut(score_0["Alley_district"], q = 9, labels=False ,duplicates="drop")
score_0["Development_district"] = pd.qcut(score_0["Development_district"], q = 9, labels=False ,duplicates="drop")
score_0["culture_space"] = pd.qcut(score_0["culture_space"], q = 9, labels=False ,duplicates="drop")
```

```python
score_0["pharmacy"] += 1
score_0["bus_station"] += 1
score_0["food"] += 1
score_0["parking"] += 1
score_0["Alley_district"] += 1
score_0["Development_district"] += 1
score_0["culture_space"] += 1
```

```python
score_0["bus_station"]
```

```python
score_0["score"] = score_0["pharmacy"]*4 + score_0["bus_station"]*3.8 + score_0["food"]*2 + score_0["parking"] + score_0["Alley_district"]*0.5 + score_0["Development_district"]*0.1 + score_0["culture_space"] * 0.01
score_0.sort_values("score", ascending=False)
```
